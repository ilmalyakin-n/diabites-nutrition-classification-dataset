```python
import pandas as pd
import numpy as np

# Jika menggunakan Google Drive, hilangkan tanda pagar pada 2 baris di bawah ini:
# from google.colab import drive
# drive.mount('/content/drive')

# 1. GATHERING DATA
print("=== 1. GATHERING DATA ===")
# Sesuaikan path file dengan lokasi file mentah kamu (contoh: nilai-gizi.csv)
file_path = '/content/drive/MyDrive/STMIK /Coding Camp/Data/Data Klasifikasi/data_mentah.csv'

try:
    df_raw = pd.read_csv(file_path)
    print(f"Berhasil memuat data mentah! Jumlah baris: {df_raw.shape[0]}, Jumlah kolom: {df_raw.shape[1]}")
    display(df_raw.head())
except FileNotFoundError:
    print(f"Error: File '{file_path}' tidak ditemukan. Pastikan file sudah di-upload ke Colab atau Drive.")
```

```python
# 2. ASSESSING DATA
print("=== 2. ASSESSING DATA ===")

# A. Cek struktur dan tipe data
print("\n[Info Dataset]")
df_raw.info()

# B. Cek jumlah data kosong (Missing Values)
print("\n[Missing Values]")
print(df_raw.isnull().sum())

# C. Cek duplikasi baris
duplikat = df_raw.duplicated().sum()
print(f"\n[Data Duplikat] Terdapat {duplikat} baris duplikat.")

# D. Cek anomali statistik dasar (misal: apakah ada nilai gizi yang negatif?)
print("\n[Statistik Deskriptif]")
display(df_raw.describe())
```

```python
# 3. CLEANING DATA
print("=== 3. CLEANING DATA ===")

# A. Filtering & Renaming (Membuang kolom tak perlu)
mapping = {
    'name': 'product_name',
    'sugar_g': 'sugar_g',
    'carbohydrate_g': 'carbs_g',
    'energy_kcal': 'calories',
    'sodium_mg': 'sodium_mg',
    'fat_g': 'fat_g'
}
df_clean = df_raw[list(mapping.keys())].rename(columns=mapping)

# B. Handling Missing Values
df_clean = df_clean.fillna(0) # Mengisi NaN dengan 0 untuk nilai gizi

# C. Handling Duplicates
df_clean = df_clean.drop_duplicates()

# D. Menghapus data anomali (Produk yang Kalori, Karbo, Lemak, dan Gula = 0 semua)
# Ini penting agar AI tidak belajar dari data "kosong" (air mineral dimaafkan jika ada nama spesifik, tapi umumnya kita hapus)
kondisi_anomali = (df_clean['calories'] == 0) & (df_clean['carbs_g'] == 0) & (df_clean['fat_g'] == 0)
df_clean = df_clean[~kondisi_anomali]

print(f"Data bersih siap digunakan. Jumlah baris saat ini: {df_clean.shape[0]}")
display(df_clean.head())
```

```python
print("=== 4. DATA AUGMENTATION (USER PROFILES) ===")

# Mendefinisikan profil user simulasi
# diabetes_type: 0 (Normal), 1 (Tipe 1), 2 (Tipe 2)
profiles = [
    {'age_group': 'adult', 'bmi_category': 'normal', 'diabetes_type': 0},
    {'age_group': 'adult', 'bmi_category': 'overweight', 'diabetes_type': 0},
    {'age_group': 'adult', 'bmi_category': 'normal', 'diabetes_type': 1},      # Tipe 1 Dewasa
    {'age_group': 'adult', 'bmi_category': 'overweight', 'diabetes_type': 2},  # Tipe 2 Overweight
    {'age_group': 'elderly', 'bmi_category': 'normal', 'diabetes_type': 2},    # Tipe 2 Lansia
    {'age_group': 'child', 'bmi_category': 'normal', 'diabetes_type': 1},      # Tipe 1 Anak
    {'age_group': 'child', 'bmi_category': 'normal', 'diabetes_type': 0}       # Anak Normal
]

augmented_rows = []

# Cross-join: 1 produk akan diuji oleh 7 profil user berbeda
for _, row in df_clean.iterrows():
    for p in profiles:
        new_row = row.copy()
        new_row['age_group'] = p['age_group']
        new_row['bmi_category'] = p['bmi_category']
        new_row['diabetes_type'] = p['diabetes_type']
        augmented_rows.append(new_row)

df_augmented = pd.DataFrame(augmented_rows)
print(f"Data setelah augmentasi: {df_augmented.shape[0]} baris.")
```

```python
print("=== 5. LABELING & EXPORT ===")

def calculate_final_label(row):
    score = 0

    # 1. Base Score (Standar Gizi)
    if row['sugar_g'] <= 5: score += 0
    elif row['sugar_g'] <= 10: score += 1
    elif row['sugar_g'] <= 20: score += 2
    else: score += 3

    if row['carbs_g'] <= 15: score += 0
    elif row['carbs_g'] <= 30: score += 1
    elif row['carbs_g'] <= 45: score += 2
    else: score += 3

    if row['calories'] <= 100: score += 0
    elif row['calories'] <= 200: score += 1
    elif row['calories'] <= 300: score += 2
    else: score += 3

    if row['sodium_mg'] <= 200: score += 0
    elif row['sodium_mg'] <= 400: score += 1
    elif row['sodium_mg'] <= 600: score += 2
    else: score += 3

    if row['fat_g'] <= 3: score += 0
    elif row['fat_g'] <= 10: score += 1
    elif row['fat_g'] <= 20: score += 2
    else: score += 3

    # 2. DIABETES TYPE MODIFIER (Sangat Ketat)
    if row['diabetes_type'] == 1:
        # Tipe 1: Sangat sensitif terhadap Karbohidrat
        score += 3
        if row['sugar_g'] > 10: score += 4
        if row['carbs_g'] > 30: score += 3
    elif row['diabetes_type'] == 2:
        # Tipe 2: Sensitif terhadap Kalori berlebih dan Lemak
        score += 3
        if row['sugar_g'] > 10: score += 4
        if row['calories'] > 200 or row['fat_g'] > 10: score += 2

    # 3. MODIFIER LAINNYA
    if row['bmi_category'] == 'overweight':
        score += 1
    if row['age_group'] == 'elderly':
        score += 1
        if row['sodium_mg'] > 300: score += 1
    if row['age_group'] == 'child' and row['sugar_g'] > 10:
        score += 2

    # 4. LABELING (3 Kategori)
    if score <= 4: return 'Recommended'
    elif score <= 9: return 'Caution'
    else: return 'Not Recommended'

# Terapkan logika ke seluruh baris
df_augmented['label'] = df_augmented.apply(calculate_final_label, axis=1)

# Simpan ke CSV Final
output_filename = 'dataset_diabites_type_3labels.csv'
df_augmented.to_csv(output_filename, index=False)

print(f"BERHASIL! Dataset disimpan sebagai '{output_filename}'\n")

print("--- DISTRIBUSI LABEL KESELURUHAN ---")
print(df_augmented['label'].value_counts())
```
